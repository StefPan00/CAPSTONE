# PULIZIA DB CONUSMO FARMACI IN ITALIA- DETTAGLIO EDUCAZIONE RICEVUTA

```python
import pandas as pd

file_path=r"Consumo_farmaci_ITA_titolostudio.csv"
df=pd.read_csv(file_path)
```

```python
df.drop_duplicates
print("shape db:", df.shape)
df.info()
```

```python
df=df.iloc[:,:-13]

df["REF_AREA"]=df["REF_AREA"].replace("IT","Italy")

Education={3:"lower primary school",
           4:"lower secondary school",
           7:"high school",
           11:"degree and post degree",
           99:"total"}

df["EDU_LEV_HIGHEST"]=df["EDU_LEV_HIGHEST"].replace(Education)

Sex={1:"Males",
     2:"Females",
     9:"Total"}

df["SEX"]=df["SEX"].replace(Sex)

Age={"Y_GE6":"<6",
     "Y_GE65":">65",
     "Y25-44":"25-44",
     "Y45-64":"45-64",
     "Y6-24":"6-24"}

df["AGE"]=df["AGE"].replace(Age)

df["FREQ"]=df["FREQ"].replace("A","Annual")

df
```

```python
#save csv 
nome_file = "CF_education_pulito.csv"
df.to_csv(nome_file, index=False) 
```

PULIZIA DATI DB: CONUSMO FARMACI- DETTAGLIO POSIZIONE LAVORATIVA

```python
#load file
file_path_età=r"Consumo_farmaci_ITA_età.csv"
df_età=pd.read_csv(file_path_età)
```

```python
#esplorazione
print("Dimensione db:",df_età.shape)
print("info db:","/n",df_età.info())
```

```python
df_età["REF_AREA"]=df_età["REF_AREA"].replace("IT","Italy")

Sex={1:"Males",
     2:"Females",
     9:"Total"}

df_età["SEX"]=df_età["SEX"].replace(Sex)

df_età["EDU_LEV_HIGHEST"]=df_età["EDU_LEV_HIGHEST"].replace(99,"Total")
df_età["LABOUR_PROFESS_STATUS_B"]=df_età["LABOUR_PROFESS_STATUS_B"].replace(99,"Total")

df_età["FREQ"]=df_età["FREQ"].replace("A","Annual")

df_età=df_età.iloc[:,:-13]


df_età["AGE"]=df_età["AGE"].str.replace("Y","",regex=False)

Age={'_GE0':"<0",
     "_GE75":">75"}

df_età["AGE"]=df_età["AGE"].replace(Age)

df_età
```

```python
nome_file_età = "CF_età_pulito.csv"
df_età.to_csv(nome_file_età, index=False) 
```

# PULIZIA DB:CONSUMO FARMACI IN ITALIA: DETTAGLIO POSIZIONE

```python
path_file_prof=r"Consumo_farmaci_professione.csv"
df_prof=pd.read_csv(path_file_prof)
```

```python
#esplorazione
print("Dimensione db:",df_prof.shape)
print("info db:","/n",df_prof.info())
```

```python
df_prof=df_prof.iloc[:,:-13]

df_prof["REF_AREA"]=df_prof["REF_AREA"].replace("IT","Italy")

Sex={1:"Males",
     2:"Females",
     9:"Total"}

df_prof["SEX"]=df_prof["SEX"].replace(Sex)

df_prof["EDU_LEV_HIGHEST"]=df_prof["EDU_LEV_HIGHEST"].replace(99,"Total")

df_prof["FREQ"]=df_prof["FREQ"].replace("A","Annual")

df_prof["AGE"]=df_prof["AGE"].replace("Y_GE15",">15")

Profession={"1":"Occupied",
            "1A":"Freelance profession",
            "1B":"Employe",
            "1C":"Trainee",
            "1D":"Self-employed worker",
            "2":"Unemployed looking for a position",
            "3":"Looking for first job",
            "11":"Unoccupied",
            "4":"Homely/Housewife",
            "5":"Student",
            "6":"Retired from work",
            "7":"Other",
            "99":"Total"}

df_prof["LABOUR_PROFESS_STATUS_B"]=df_prof["LABOUR_PROFESS_STATUS_B"].replace(Profession)
df_prof
```

```python
nome_file_prof = "CF_profession_pulito.csv"
df_prof.to_csv(nome_file_prof, index=False) 
```
